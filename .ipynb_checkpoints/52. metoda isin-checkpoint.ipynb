```python
import pandas as pd
```

```python
employees = pd.read_csv('employees.csv')
```

```python
sales = employees['department'] == 'Sales'
engineering = employees.department == 'Engineering'
product_management = employees.department == 'Product Management'
employees[sales | engineering | product_management]
```

```python
departments = sales | engineering | product_management
employees[departments]
```

```python
employees['department'].isin(['Sales', 'Engineering', 'Product Management'])
employees[employees['department'].isin(['Sales', 'Engineering', 'Product Management'])]
```
